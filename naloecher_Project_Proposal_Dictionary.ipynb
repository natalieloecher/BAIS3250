```python
import pandas as pd
```

```python
weather_df = pd.read_csv("Chicago Illinois 2023-04-01 to 2025-04-01.csv", encoding="utf-8", sep=",")
```

```python
display(weather_df)
```

```python
weather_df['conditions'].unique()
```

```python
def simplify_weather(condition):
    condition = condition.lower()
    if 'snow' in condition or 'ice' in condition or 'sleet' in condition:
        return 'Snow'
    elif 'rain' in condition or 'drizzle' in condition:
        return 'Rain'
    elif 'cloudy' in condition or 'overcast' in condition:
        return 'Cloudy'
    elif 'clear' in condition or 'sunny' in condition or 'fair' in condition:
        return 'Sunny'
    else:
        return 'Cloudy'
weather_df['conditions'] = weather_df['conditions'].apply(simplify_weather)
```

```python
print(weather_df.head())
```

```python
print(weather_df.dtypes)
```

```python
weather_df = weather_df.drop(columns=["feelslikemax", "feelslikemin", "dew", "precipcover", "precipprob", "preciptype", "windgust", "winddir", "sealevelpressure", "solarradiation", "solarenergy", "sunrise", "sunset", "moonphase", "description", "icon", "stations", ])


weather_df.head()
```

```python
weather_df.rename(columns={
    "name": "City",
    "datetime": "Date",
    "tempmax": "Temp Max (°F)",
    "tempmin": "Temp Min (°F)",
    "temp": "Avg Temp (°F)",
    "feelslike": "Feels Like Temp (°F)", 
    "humidity": "Humidity (%)",
    "precip": "Precipitation (in)",
    "snow": "Snow (in)",
    "snowdepth": "Snow Depth (in)",
    "windspeed": "Wind Speed (mph)",
    "cloudcover": "Cloud Cover (%)",
    "visibility": "Visibility (mi)",
    "uvindex": "UV Index",
    "severerisk": "Severe Weather Risk",
    "conditions": "Conditions"
}, inplace=True)

weather_df.head()
```

```python
weather_df.to_csv("weather_data.csv", index=False)
```

```python
weather_dict = pd.DataFrame({
    "Field": weather_df.columns,
    "Type": [weather_df[col].dtype for col in weather_df.columns],
    "Description": ["description" for col in weather_df.columns]  # You’ll need to manually fill this part
})


print(weather_dict)
```

```python
weather_dict["Description"] = [
    "Location",   # For 'name'
    "Date of the record (YYYY,MM,DD)",      # For 'datetime'
    "Maximum temperature that day (°F)",            # For 'tempmax'
    "Minimum temperature that day (°F)",              # For 'tempmin'
    "Temperature that day (°F)",            # For 'temp'
    "'Feels Like' temperature that day (°F)",  # For 'feelslike'
    "Humidity percentage in the air", # For 'humidity'
    "Precipitation amount in inches",            # For 'precip'
    "Snow amount in inches",  # For 'snow'
    "Snow depth on ground in inches",            # For 'snowdepth'
    "Wind speed that day (mph)",  # For 'windspeed'
    "Percentage of cloud covered sky",            # For 'cloudcover'
    "Distance that can be seen in miles",  # For 'visibility'
    "Strength of ultraviolet radiation from the sun",            # For 'uvindex'
    "Chance of severe weather",  # For 'severerisk'
    "General weather condition (ex, Sunny, Rainy)"  # For 'conditions'
]

print(weather_dict)
```
